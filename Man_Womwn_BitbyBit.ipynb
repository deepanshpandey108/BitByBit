```python
import zipfile
import os
```

```python
uploaded_zip = '/content/Gender Detection and Labelling.v1i.yolov11-20260207T101936Z-1-001.zip'
extracted_path = '/content/dataset'
```

```python
with zipfile.ZipFile(uploaded_zip, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

print("âœ… Dataset extracted to:", extracted_path)
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")  # small & fast

results = model.train(
    data="/content/dataset/Gender Detection and Labelling.v1i.yolov11/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    patience=10,
    lr0=0.01
)
```
